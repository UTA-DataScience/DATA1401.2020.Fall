```python
checkers.size=10
```

```python
print(checkers.size)
```

```python
checkers.get_size()
```

In general this isn't good practice. But it does illustrate an important python behavior. 

'import` does not load a library again if it has already been loaded:

```python
import checkers
```

```python
print(checkers.size)
```

```python
checkers.get_size()
```

```python
import importlib
importlib.reload(checkers)
checkers.size
```
